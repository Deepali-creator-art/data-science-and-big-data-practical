3. Handle Missing Numeric Values (e.g., Age)

ðŸ‘‰ Use mean, median, or model-based imputation.

```python
from sklearn.impute import SimpleImputer

# Impute 'age' with median
imputer = SimpleImputer(strategy='median')
df['Age'] = imputer.fit_transform(df[['Age']])
df
```

```python
df.isnull().sum()
```

4. Handle Missing Categorical Values (e.g., Embarked)

ðŸ‘‰ Use most frequent value or constant.

```python
df['Embarked'].value_counts()
```

```python
nmg = pd.notnull(df["Embarked"])

nmgd= df[nmg]
print(nmgd)
```

```python
df[df['Embarked'].isnull()]
```

```python
# Impute 'embarked' with most frequent
cat_imputer = SimpleImputer(strategy='most_frequent')
df['Embarked'] = cat_imputer.fit_transform(df[['Embarked']]).ravel()
```

```python
num_cols = df.select_dtypes(include=np.number).columns
print(num_cols.tolist())
```

```python
from scipy.stats import zscore
#parameter nan_policy = 'omit' and nans are ignored in the calculation

z_scores = df[num_cols].apply(zscore, nan_policy="omit")
print(z_scores)
# Flag anything with |z| â‰¥ 3
mask_z = (np.abs(z_scores) >= 3)

out_z = (mask_z.sum()
         .reset_index()
         .rename(columns={"index": "column", 0: "n_outliers"}))
print(out_z.sort_values("n_outliers", ascending=False))
```
